```python
import os
import argparse
import logging
import sys

import torch
from torch.optim.lr_scheduler import StepLR
import torchtext

sys.path.append(os.path.dirname(os.path.abspath(os.path.dirname(os.path.abspath(os.path.dirname('__file__'))))))

from trainer.supervised_trainer import SupervisedTrainer
from trainer.supervised_trainer_correction import SupervisedTrainer_correction
from models.encoderRNN_lenemb_add import EncoderRNN_lenemb_add
from models.decoderRNN_lenemb_add import DecoderRNN_lenemb_add
from models.seq2seq import Seq2seq
from loss.loss import Perplexity
from optim.optim import Optimizer
from dataset import fields
from evaluator.predictor import Predictor

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.decomposition import PCA
import numpy as np

from sklearn.manifold import TSNE
from collections import deque
import math
```

```python
train_path = "../../data/copy_palindrome_concat/correction_separator_Ctype4_concat_N100/data_train.txt"
log_path = "../../log/pth/copy_palindrome_correction_separator_Ctype4_concat2_lenemb_add_separator_Ctype4_concat_N100_model_save.pth"
```

```python
LOG_FORMAT = '%(asctime)s %(name)-12s %(levelname)-8s %(message)s'
logging.basicConfig(format=LOG_FORMAT, level=getattr(logging, "info".upper()))
```

```python
# Prepare dataset
src = fields.SourceField()
tgt = fields.TargetField()
max_len = 104
def len_filter(example):
    return len(example.src) <= max_len and len(example.tgt) <= max_len
train = torchtext.data.TabularDataset(
    path=train_path, format='tsv',
    fields=[('src', src), ('tgt', tgt)],
    filter_pred=len_filter
)
```

```python
src.build_vocab(train)
tgt.build_vocab(train)
input_vocab = src.vocab
output_vocab = tgt.vocab
```

```python
# Prepare loss
weight = torch.ones(len(tgt.vocab))
pad = tgt.vocab.stoi[tgt.pad_token]
loss = Perplexity(weight, pad)
if torch.cuda.is_available():
    loss.cuda()

seq2seq = None
seq2seq_no = None
optimizer = "Adam"
hidden_size = 200
bidirectional = False
```

# Initialize model Luong att

```python
# Initialize model
encoder = EncoderRNN_lenemb_add(len(src.vocab), max_len, hidden_size, n_layers=1,
                    bidirectional=bidirectional, variable_lengths=True, context_vector=True)
decoder = DecoderRNN_lenemb_add(len(tgt.vocab), max_len, hidden_size * 2 if bidirectional else hidden_size,
                    dropout_p=0.2, use_attention="Luong", bidirectional=bidirectional, n_layers=1,
                    eos_id=tgt.eos_id, sos_id=tgt.sos_id)
seq2seq = Seq2seq(encoder, decoder)
if torch.cuda.is_available():
    seq2seq.cuda()

for param in seq2seq.parameters():
    param.data.uniform_(-0.08, 0.08)

seq2seq.load_state_dict(torch.load(log_path))
seq2seq.eval()

predictor = Predictor(seq2seq, input_vocab, output_vocab)
```

# Answer function

```python
def answer_function(data):
    results = []
    words = data.split("#")
    queue = deque()
    for word in words:
        result = ""
        m = int(len(word)/2)
        for i in range(0,m):
            result += "0"
            queue.append(word[i])
        if len(word)%2 != 0:
            result += "0"
            m += 1
        for i in range(m,len(word)):
            w = queue.popleft()
            if word[i] != w:
                if w == "a":
                    result += "1"
                elif w == "b":
                    result += "2"
                elif w == "c":
                    result += "3"
                elif w == "d":
                    result += "4"
                else:
                    result += "0"
            else:
                result += "0"
                
        results.append(result)

    return " ".join("0".join(results))
```

# Step function

```python
def step_function(x):
    
    if x > 0:
        if x < 0.5:
            return 0
        else:
            return 1
    else:
        if x > -0.5:
            return 0
        else:
            return -1
```
