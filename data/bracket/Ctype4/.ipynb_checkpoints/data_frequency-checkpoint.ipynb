# Import packages

import useful packages for experiments

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
```

```python
lines = open('grammar_data_N100.txt').read().strip().split('\n')
```

- depth

```python
pairs = [[s for s in l.split('\t')] for l in lines]
depths = []
for pair in tqdm(pairs):
    depth = []
    for i in pair[1].split(' '):
        depth.append(int(i))
    depths.append(max(depth))
```

- distance

```python
distance = []
for pair in tqdm(pairs):
    reverse_pair = list(reversed((pair[1].split(' '))))
    reverse_pair.append('0')
    depth = []
    count = -2
    compare = reverse_pair[0]
    for i in reverse_pair:
        if(i == compare):
            count = count + 1
            depth.append(count)
            count = -2
        else:
            count = count + 1
    depth.append(count)
    distance.append(max(depth))
```

- length

```python
lengths = []
for pair in tqdm(pairs):
    length = 0
    for i in pair[1].split(' '):
        length += 1
    lengths.append(length)
```

# Experimental results

- depth to frequency

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.hist(depths, bins=10)
plt.xlabel('depth', fontsize=20)
plt.ylabel('frequency', fontsize=20)
plt.title('depth to frequency', fontsize=26)
plt.savefig('depth_to_frequency.png')
```

- distance to frequency

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.hist(distance, bins=10)
plt.xlabel('distance', fontsize=20)
plt.ylabel('frequency', fontsize=20)
plt.title('distance to frequency', fontsize=26)
plt.savefig('distance_to_frequency.png')
```

- length to frequency

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.hist(lengths, bins=10)
plt.xlabel('length', fontsize=20)
plt.ylabel('frequency', fontsize=20)
plt.title('length to frequency', fontsize=26)
plt.savefig('length_to_frequency.png')
```
